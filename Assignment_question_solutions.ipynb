##### Question 16: How to perform train, validation, and test splits on the weather data?
##### Note: 80% training data, In the remaining 20% again take 75% for validation and 25% for testing

```python
from sklearn.model_selection import train_test_split

# First split: 80% train, 20% temp (validation + test)
train_data, temp_data = train_test_split(weather_data, test_size=0.20, random_state=42)

# Second split: 75% validation, 25% test of the temp set (20%)
validation_data, test_data = train_test_split(temp_data, test_size=0.25, random_state=42)

# Display the shapes of the resulting datasets
print("Train data shape:", train_data.shape)
print("Validation data shape:", validation_data.shape)
print("Test data shape:", test_data.shape)
```

###### Question 17: How to identify the missing values in the dataset?

```python
# Identify missing values
missing_values = weather_data.isnull().sum()

# Display the number of missing values in each column
print("Missing values in each column:")
print(missing_values)
```

##### Count the number of numerical columns with missing values and categorical columns with missing values.

```python
# Count the number of numerical columns with missing values
num_missing_numerical = weather_data.select_dtypes(include=['float64', 'int64']).isnull().sum()
num_missing_numerical_count = num_missing_numerical[num_missing_numerical > 0].count()

# Count the number of categorical columns with missing values
num_missing_categorical = weather_data.select_dtypes(include=['object']).isnull().sum()
num_missing_categorical_count = num_missing_categorical[num_missing_categorical > 0].count()

print("Number of numerical columns with missing values:", num_missing_numerical_count)
print("Number of categorical columns with missing values:", num_missing_categorical_count)
```

###### For explanation purpose, consider 16 numerical columns and divide them into two sets, each containing 8 columns. Impute the mean for the first set of 8 columns and impute the median for the second set of 8 columns. Use mode imputation for the categorical columns.

```python
# Divide numerical columns into two parts
numerical_cols = train_data.select_dtypes(include=['float64', 'int64']).columns
num_cols_mean = numerical_cols[:len(numerical_cols)//2]  # First half
num_cols_mode = numerical_cols[len(numerical_cols)//2:]  # Second half

# Categorical columns
categorical_cols = train_data.select_dtypes(include=['object']).columns
```

```python
numerical_cols
```

##### Question 18: On the train, validation, test sets perform the mean, median and mode imputations.

```python
# Impute numerical columns with mean in training data
for col in num_cols_mean:
    train_data[col].fillna(train_data[col].mean(), inplace=True)
    validation_data[col].fillna(train_data[col].mean(), inplace=True)  # Use train mean for validation data
    test_data[col].fillna(train_data[col].mean(), inplace=True)  # Use train mean for test data
```

```python
# Impute numerical columns with mode in training data
for col in num_cols_mode:
    train_data[col].fillna(train_data[col].mode()[0], inplace=True)
    validation_data[col].fillna(train_data[col].mode()[0], inplace=True)  # Use train mode for validation data
    test_data[col].fillna(train_data[col].mode()[0], inplace=True)  # Use train mode for test data
```

```python
# Impute categorical columns with mode in training data (more typical than median)
for col in categorical_cols:
    train_data[col].fillna(train_data[col].mode()[0], inplace=True)
    validation_data[col].fillna(train_data[col].mode()[0], inplace=True)  # Use train mode for validation data
    test_data[col].fillna(train_data[col].mode()[0], inplace=True)  # Use train mode for test data
```

##### Question 19: How to show that there are no missing values after imputation?

```python
# Check for missing values in the imputed dataframes
missing_values_train = train_data.isnull().sum()
missing_values_validation = validation_data.isnull().sum()
missing_values_test = test_data.isnull().sum()

print("Missing values in training data after imputation:", missing_values_train)
print("Missing values in validation data after imputation:", missing_values_validation)
print("Missing values in test data after imputation:", missing_values_test)
```

##### Question 20: perform standardization on two numerical columns {MinTemp, MaxTemp}

```python
from sklearn.preprocessing import StandardScaler

# Select two numerical columns to standardize
columns_to_standardize = ['MinTemp', 'MaxTemp']

# Initialize the StandardScaler
scaler = StandardScaler()

# Standardize the selected columns in the training data
train_data_std = train_data.copy()
train_data_std[columns_to_standardize] = scaler.fit_transform(train_data[columns_to_standardize])

# Standardize the selected columns in the validation data using the same scaler
validation_data_std = validation_data.copy()
validation_data_std[columns_to_standardize] = scaler.transform(validation_data[columns_to_standardize])

# Standardize the selected columns in the test data using the same scaler
test_data_std = test_data.copy()
test_data_std[columns_to_standardize] = scaler.transform(test_data[columns_to_standardize])
```

##### This part of the code compares the standardized columns with actual column values in training data

```python
# Compare the standardized columns with the actual columns in the training data
comparison_train = pd.DataFrame({
    'Actual MinTemp': train_data['MinTemp'],
    'Standardized MinTemp': train_data_std['MinTemp'],
    'Actual MaxTemp': train_data['MaxTemp'],
    'Standardized MaxTemp': train_data_std['MaxTemp']
})

# Display the comparisons
print("Training Data Comparison:")
print(comparison_train.head())
```

##### Question 22: perform label encoding on a categorical column 'WindGustDir'

```python
from sklearn.preprocessing import LabelEncoder

# Select a categorical column to perform label encoding
categorical_column = 'WindGustDir'

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Perform label encoding on the selected column in the training data
train_data_encoded = train_data.copy()
train_data_encoded[categorical_column] = label_encoder.fit_transform(train_data[categorical_column].astype(str))

# Perform label encoding on the selected column in the validation data using the same encoder
validation_data_encoded = validation_data.copy()
validation_data_encoded[categorical_column] = label_encoder.transform(validation_data[categorical_column].astype(str))

# Perform label encoding on the selected column in the test data using the same encoder
test_data_encoded = test_data.copy()
test_data_encoded[categorical_column] = label_encoder.transform(test_data[categorical_column].astype(str))
```

###### Question 23: perform one hot encoding on a categorical column 'WindDir3pm'

```python
from sklearn.preprocessing import OneHotEncoder
# Select the column to perform one-hot encoding on
column_to_encode = 'WindDir3pm'

# Initialize the OneHotEncoder
encoder = OneHotEncoder(sparse=False, drop='first')

# Fit the encoder on the training data and transform the column
train_encoded = encoder.fit_transform(train_data[[column_to_encode]])
validation_encoded = encoder.transform(validation_data[[column_to_encode]])
test_encoded = encoder.transform(test_data[[column_to_encode]])

# Create DataFrames for the encoded columns with appropriate column names
train_encoded_df = pd.DataFrame(train_encoded, columns=encoder.get_feature_names_out([column_to_encode]))
validation_encoded_df = pd.DataFrame(validation_encoded, columns=encoder.get_feature_names_out([column_to_encode]))
test_encoded_df = pd.DataFrame(test_encoded, columns=encoder.get_feature_names_out([column_to_encode]))

# Reset indices to concatenate with the original DataFrames
train_encoded_df.reset_index(drop=True, inplace=True)
validation_encoded_df.reset_index(drop=True, inplace=True)
test_encoded_df.reset_index(drop=True, inplace=True)
```

##### This part of the code compares the one hot encoded categorical column with actual column values in training data

```python
# Concatenate the original and encoded DataFrames for comparison
comparison_train_encoded = pd.concat([train_data[[column_to_encode]].reset_index(drop=True), train_encoded_df], axis=1)
# Display the comparisons
print("Training Data One-Hot Encoding Comparison:")
print(comparison_train_encoded.head())
```
